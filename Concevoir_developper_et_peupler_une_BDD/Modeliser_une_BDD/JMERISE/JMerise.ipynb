---

Ce tableau n'est pas adéquat car il faut créer une colonne pour chaque enfant supplémentaire par employé. De plus il n'y a pas de clé primaire apparente (si c'est le nom/prénom de l'employé, on ne pourra pas avoir deux employés au même nom). Aussi on séparera le Nom/prénom en deux champs.<br>
<br>
Il serait préférable de créer deux tables, employés et enfants, et d'ajouter des colonnes "ID" à chaque table qui serra la clé primaire. On connectera ces tables où un employé peut avoir plusieurs enfants et un(des) enfant(s) ne peut appartenir qu'à un employé. <br>
<br>
On peut aussi ajouter que l'adresse n'est pas atomique, on créée donc une tables "Adresse" où le numéro de rue, le nom de rue, le code postal et la ville sont des champs séparés.

## Exercice 4

Une bibliothèque de prêt dispose d'un certain nombre d'ouvrages, classés par rayon (Littérature, Histoire, Géographie, etc.). Chaque ouvrage est l'œuvre d'un ou plusieurs auteurs, et doit également être référencé selon un certain nombre de mots-clés. <br>
<br>
Chaque adhérent peut emprunter jusqu'à 5 livres en même temps et est caractérisé par la date d’emprunt.<br>
<br>
Un auteur peut être référencé dans la base même s’il n’est pas relié à un ouvrage. En revanche, un ouvrage est forcément rattaché à un auteur. Un rayon peut être vide mais un livre est forcément classé dans un rayon. <br>
<br>
Un mot clé peut exister sans être rattaché à un ouvrage en revanche un ouvrage ne peut pas exister sans être référencé par au moins un mot clé. On se place dans deux cas successifs, de complexité croissante : <br>
<br>
### 1) la bibliothèque ne possède qu'un seul exemplaire de chaque ouvrage. Elle enregistre uniquement les emprunts présents (il n'y a pas d'historique des emprunts passés).

![alternative text](Images/Exercice_4_1.jpg)

---

### 2) la bibliothèque ne possède qu'un seul exemplaire de chaque ouvrage, mais elle tient un historique de tous les emprunts qui ont été effectués. Pour réaliser cela, il faut donc rajouter la date de retour en plus de la date d’emprunt.

## Exercice 5

Un historien souhaite établir des statistiques sur des soldats de la Première Guerre mondiale. <br>
<br>
Chaque soldat est identifié par son nom, son prénom, sa date de naissance, son état civil et sa date de décès (qui peut être nulle si le soldat n’est pas décédé). <br>
<br>
Pour chaque soldat, on identifiera le type des blessures qu’il a reçues (type et date de la blessure) ainsi que la ou les batailles auxquelles il a participé. Les batailles seront référencées par leur nom, leur lieu, ainsi que les dates de début et de fin. <br>
<br>
Pour chaque soldat, on identifiera son grade le plus élevé et sa date de promotion à ce grade. Chaque grade est identifié par l’intitulé du grade et l’échelon (1, 2, 3, etc.). <br>
<br>
Chaque soldat a au moins un grade. 
<br>
Enfin, on identifiera toutes les unités auxquelles le soldat a appartenu. L’unité est définie par son intitulé. Les dates d’affectation aux unités sont également référencées pour chaque soldat.<br> 
<br>
Établir le MCD et le MLD adéquats avant de les tester dans MariaDB.

## Exercice 10

On cherche à stocker dans une base de données la structure administrative du territoire français. Celle-ci se décompose en plusieurs étages. <br>
<br>
Dans l'ordre : la commune, le canton, l'arrondissement, le département, la région. A chaque fois, un étage ne peut être rattaché qu'à un seul étage supérieur, tandis que l'étage supérieur contient plusieurs éléments de l'étage inférieur.<br>
Chaque étage à partir du canton possède une commune « capitale », qui est obligatoirement chef de tous les niveaux inférieurs (on ne peut pas avoir une capitale de région qui ne soit aussi capitale de département, d'arrondissement et de canton).<br>
<br>
On ajoute à cela la question des contiguïtés territoriales des communes : on veut savoir, pour chaque commune, avec quelles autres communes elle possède une limite commune.<br>
<br>
Etablir le MCD et le MLD adéquat.
