# Trier des fichiers

## Notions abordées

- Boucle et Condition
- Fonction (et récursivité)
- Liste
- Module ```os```
- Module ```shutil```

## Énoncé

Vous devez créer une fonction qui trie automatiquement des fichiers dans des sous-dossiers en fonction de leur type (et donc de leur extension).
Facultatif : Réaliser la fonction inverse → déplacer les fichiers des sous-dossiers dans le dossier principale.

Exemple :
```
├── facture.pdf 
├── Tour_eiffel.jpg     
├── Logo.png    
├── Despacito.mp3 
├── Court_metrage.mp4  
```

Résultat :
```
├── Documents
|   ├── Facture.pdf 
├── Images  
│   ├── Tour_eiffel.jpg     
│   ├── Logo.png    
├── Musiques
│   ├── Despacito.mp3 
├── Videos    
│   ├── Court_metrage.mp4        
```

````python
import os
import shutil

try:
    os.chdir(os.getcwd()+'\\Fichiers')
except FileNotFoundError:
    print('Les fichiers ont disparus !')

def trier_fichiers() :
    listeFichier = os.listdir()
    toRemove=[]
    for i in range (0,len(listeFichier)):
        if os.path.isdir(listeFichier[i]):
            toRemove.append(listeFichier[i])
    for i in range (0,len(toRemove)):
        listeFichier.remove(toRemove[i])    

    try:        
        os.mkdir('Images')
        os.mkdir('Musiques')
        os.mkdir('Vidéos')
        os.mkdir('Documents')
        os.mkdir('Inclassés')
    except:
        pass
    
    for i in range(0,len(listeFichier)):
        if '.mp3' in listeFichier[i][-4:]:
            try:
                shutil.move(listeFichier[i],"Musiques")
            except:
                pass
        
        elif '.mp4' in listeFichier[i][-4:]:
            try:
                shutil.move(listeFichier[i],"Vidéos")
            except:
                pass
        
        elif '.pdf' in listeFichier[i][-4:] or '.txt' in listeFichier[i][-4:]:
            try:
                shutil.move(listeFichier[i],"Documents")
            except:
                pass
    
        elif ('.png' in listeFichier[i][-4:]) or ('.jpg' in listeFichier[i][-4:]):
            try:
                shutil.move(listeFichier[i],"Images")
            except:
                pass
        
        else:
            try:
                shutil.move(listeFichier[i],"Inclassés")
            except:
                pass
        
def detrier_fichiers():
    try:
        os.chdir(os.getcwd()+'\\Documents')
        listeFichier = os.listdir()
        for i in range (0,len(listeFichier)):
            shutil.move(listeFichier[i],'..')
        os.chdir('..')
    except:
        pass
    
    try:
        os.chdir(os.getcwd()+'\\Vidéos')
        listeFichier = os.listdir()
        for i in range (0,len(listeFichier)):
            shutil.move(listeFichier[i],'..')
        os.chdir('..')
    except:
        pass
    
    try:
        os.chdir(os.getcwd()+'\\Images')
        listeFichier = os.listdir()
        for i in range (0,len(listeFichier)):
            shutil.move(listeFichier[i],'..')
        os.chdir('..')
    except:
        pass
    
    try:
        os.chdir(os.getcwd()+'\\Musiques')
        listeFichier = os.listdir()
        for i in range (0,len(listeFichier)):
            shutil.move(listeFichier[i],'..')
        os.chdir('..')
    except:
        pass
    
    try:
        os.chdir(os.getcwd()+'\\Inclassés')
        listeFichier = os.listdir()
        for i in range (0,len(listeFichier)):
            shutil.move(listeFichier[i],'..')
        os.chdir('..')
    except:
        pass
    
    try:
        shutil.rmtree(os.getcwd()+'\Documents')
        shutil.rmtree(os.getcwd()+'\Musiques')
        shutil.rmtree(os.getcwd()+'\Vidéos')
        shutil.rmtree(os.getcwd()+'\Images')
        shutil.rmtree(os.getcwd()+'\Inclassés')
    except:
        pass

while True:
    try:
        menu = int(input('Voulez vous trier vos fichiers (1), déclasser vos fichiers (2), ou quitter le programme (0)?\n'))
    except:
        print('Entrée invalide!\`n')
    
    if menu == 1:
        trier_fichiers()
    elif menu == 2:
        detrier_fichiers()
    elif menu == 0:
        break
````
