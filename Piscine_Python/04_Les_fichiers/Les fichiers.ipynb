# Exercice 1:

Ecrivez un programme qui permet à l'utilisateur d'écrire un fichier texte simple (.txt). 
Le programme demande à l'utilisateur le nom du fichier, ainsi que son contenu. 
Après validation, le fichier est enregistré sur l'ordinateur. 


````python
import os

while True:
    
    if os.getcwd()[-8:] != 'Fichiers':
        try:
            os.chdir(os.getcwd()+'\\Fichiers')
            break
        except FileNotFoundError:
            os.mkdir('Fichiers')
    else:
        break
        
def ecrire_fichier():
    while True:
        nom = str(input("Entrez le nom du fichier: "))
        contenu = str(input("Entrez le contenu du fichier {}.txt: ".format(nom)))

        fichier = open("{}.txt".format(nom), "w")
        fichier.write(contenu)
        fichier.close()
        
        continuer = input("Tapez oui si vous voulez créer un autre fichier: ")
        if continuer != 'oui':
            break
            
ecrire_fichier()
os.chdir('..')
````

# Exercice 2:

Ecrivez un programme qui permet à l'utilisateur de lire un fichier texte simple (.txt). Le programme demande à l'utilisateur le nom du fichier. Après validation, le fichier est ouvert et son contenu est affiché.

````python
import os

while True:
    
    if os.getcwd()[-8:] != 'Fichiers':
        try:
            os.chdir(os.getcwd()+'\\Fichiers')
            break
        except FileNotFoundError:
            os.mkdir('Fichiers')
    else:
        break
        
def lire_fichier():
    
    while True:
        nom = str(input("Quel est le nom du fichier à ouvrir ?"))
        
        if nom+'.txt' not in os.listdir():
            print("Votre fichier est introuvable !")
            
        else:
            fichier = open("{}.txt".format(nom), "r")
            
            print("Contenu de votre fichier {}.txt:\n\n".format(nom),fichier.read(),'\n\n')
            fichier.close()
        
        continuer = input("Tapez oui si vous voulez lire un autre fichier: ")
        if continuer != 'oui':
            break
    
lire_fichier()
os.chdir('..')
````

# Exercice 3:

Écrivez un programme qui ouvre un fichier texte, et crée un nouveau fichier avec le contenu du premier en triplant tous les espaces entre les mots.

````python
import os
import re

while True:
    
    if os.getcwd()[-8:] != 'Fichiers':
        try:
            os.chdir(os.getcwd()+'\\Fichiers')
            break
        except FileNotFoundError:
            os.mkdir('Fichiers')
    else:
        break
        
def tripler_espace():
    
    while True:
        nom = str(input("Quel est le nom du fichier où vous voulez tripler les espaces ?"))
        
        if nom+'.txt' not in os.listdir():
            print("Votre fichier est introuvable !")
            
        else:
            fichier = open("{}.txt".format(nom), "r")
            contenu = fichier.read()
            fichier.close()
            
            wordList = re.sub("[^\w]", " ", contenu).split()
            print(wordList)
            contenu = ''
            for i in range(0,len(wordList)):
                contenu += wordList[i] + '   '
            
            fichier = open("{}TripleEspace.txt".format(nom), "w")
            fichier.write(contenu)
            fichier.close()
        
        continuer = input("Tapez oui si vous voulez tripler les espaces d'un autre fichier :")
        if continuer != 'oui':
            break
    
tripler_espace()
os.chdir('..')
````

# Exercice 4:

Écrire un programme qui ouvre deux fichiers texte et qui en crée un troisième comportant le contenu des deux premiers à la suite. <br />Bis : mettre ce programme dans une fonction que l'on pourra appeler en précisant en paramètres le chemin relatif des deux fichiers à lire ainsi que celui à écrire.

````python
import os

while True:
    
    if os.getcwd()[-8:] != 'Fichiers':
        try:
            os.chdir(os.getcwd()+'\\Fichiers')
            break
        except FileNotFoundError:
            os.mkdir('Fichiers')
    else:
        break

def combiner_texte (chemin1,chemin2,chemin3):
    
    while True:
        cheminCourant = os.getcwd()
        os.chdir(chemin1)
        print("Votre premier dossier contient:",os.listdir(),'\n')
        selection1 = str(input("Quel fichier voulez vous combiner?(ne pas mettre l'extension)\n"))
        try:                
            fichier = open("{}.txt".format(selection1), "r")
            contenu1 = fichier.read()
            fichier.close()
            os.chdir(cheminCourant)
            break
        except:
            print("Ce fichier n'existe pas!")
            
    while True:
        cheminCourant = os.getcwd()
        os.chdir(chemin1)
        print("Votre deuxième dossier contient:",os.listdir(chemin2),'\n')
        selection2 = str(input("Quel fichier voulez vous combiner avec le premier ?(ne pas mettre l'extension)\n"))
        try:                
            fichier = open("{}.txt".format(selection2), "r")
            contenu2 = fichier.read()
            fichier.close()
            os.chdir(cheminCourant)
            break
        except:
            print("Ce fichier n'existe pas!")
            

        
    cheminCourant = os.getcwd()
    os.chdir(chemin3)
    fichier = open("{}+{}.txt".format(selection1,selection2), "w")
    contenu1 = fichier.write(contenu1)
    fichier.close()
    fichier = open("{}+{}.txt".format(selection1,selection2), "a")
    fichier.write("\n"+contenu2)
    fichier.close()
    os.chdir(cheminCourant)

chemin = os.getcwd()       
combiner_texte (chemin,chemin,chemin)
os.chdir('..')
````

# Exercice 5:

Écrire un programme permettant d'ouvrir un fichier texte et de donner le nombre d'apparitions d'une liste de mots. 
<br />Bis : mettre ce programme dans une fonction qui prendra en paramètres le nom du fichier à ouvrir et la liste des mots à chercher. 
<br />Ter : le programme renvoie un objet JSON contenant le résultat de la recherche. 
<br />Par exemple : 
```json
{
  premier_mot: 2,
  deuxieme_mot: 3, 
  ...
}
```

````python
import os
import re
import json

while True:
    
    if os.getcwd()[-8:] != 'Fichiers':
        try:
            os.chdir(os.getcwd()+'\\Fichiers')
            break
        except FileNotFoundError:
            os.mkdir('Fichiers')
    else:
        break

def compter_apparition (texte,listeMot):
    
    apparition = {}
    
    fichier = open("{}.txt".format(texte), "r")
    contenu = fichier.read()
    fichier.close()
        
    wordList = re.sub("[^\w]", " ", contenu).split()
    for i in range (0,len(listeMot)):
        apparition[listeMot[i]] = 0
    
    for i in range (0,len(listeMot)):
        for x in range (0,len(wordList)):
            if wordList[x].lower() == listeMot[i]:
                apparition[listeMot[i]] += 1
    
    with open('Comptage.json', 'w') as fichier:
        fichier.write(json.dumps(apparition))
    
    
compter_apparition('Biographie_Aristote',['le','de','un','et','il','avoir','ne'])              
        
os.chdir('..')        
````
