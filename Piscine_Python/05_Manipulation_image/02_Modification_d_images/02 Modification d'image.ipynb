## 2.1 Passage en niveau de gris

```python
import os

while True:
    
    if os.getcwd()[-6:] != 'Images':
        try:
            os.chdir(os.getcwd()+'\\Images')
            break
        except FileNotFoundError:
            os.mkdir('Images')
    else:
        break

from PIL import Image

joconde = Image.open("la joconde 600.jpg")

larg, haut = joconde.size

im = Image.new("RGB",(larg,haut),"black")

for y in range (0,haut):
    for x in range (0,larg):
        r,g,b = joconde.getpixel ((x,y))
        gris = round((r+g+b)/3)
        im.putpixel((x,y),(gris,gris,gris))
im.show()
im.save("Joconde_Niveau_De_Gris.png","png")
        
```

## 2.6. Pixellisation 5x5 de l’image


```python
import os

while True:
    
    if os.getcwd()[-6:] != 'Images':
        try:
            os.chdir(os.getcwd()+'\\Images')
            break
        except FileNotFoundError:
            os.mkdir('Images')
    else:
        break

from PIL import Image

joconde = Image.open("la joconde 600.jpg")

larg, haut = joconde.size

im = Image.new("RGB",(larg,haut),"black")
grosPixel={}
for y in range (2,haut,4):
    for x in range (2,larg,4):
        try :
            r,g,b = 0,0,0
            grosPixel[x,y]={}
            for y2 in range (y-2,y+2):
                for x2 in range (x-2,x+2):
                    grosPixel[x,y][x2,y2] = joconde.getpixel((x2,y2))
            for item in grosPixel[x,y]:
                r += grosPixel[x,y][item][0]
                g += grosPixel[x,y][item][1]
                b += grosPixel[x,y][item][2]
            r = round(r/25)
            g = round(g/25)
            b = round(b/25)
            for y2 in range (y-2,y+2):
                for x2 in range (x-2,x+2):
                    im.putpixel((x2,y2),(r,g,b)) 
        except IndexError:
            pass
        
im.show()
im.save("Joconde_Pixellisation_5x5.png","png")
```

## 3.1 Ecrire une fonction qui permet le passage en noir et blanc d’une image

```python
import os

while True:
    
    if os.getcwd()[-6:] != 'Images':
        try:
            os.chdir(os.getcwd()+'\\Images')
            break
        except FileNotFoundError:
            os.mkdir('Images')
    else:
        break

from PIL import Image

def passage_noir_blanc (im,seuil):

    larg, haut = im.size

    imNB = Image.new("RGB",(larg,haut),"black")

    for y in range (0,haut):
        for x in range (0,larg):
            r,g,b = im.getpixel ((x,y))
            couleur = round((r+g+b)/3)
            if couleur < seuil:
                imNB.putpixel((x,y),(0,0,0))
            elif couleur >= seuil:
                imNB.putpixel((x,y),(255,255,255))
    return (imNB)

while True:
    while True:
        
        listeFichier = os.listdir()
        toRemove=[]
        for i in range (0,len(listeFichier)):
            if listeFichier[i][-4:] != '.jpg':
                toRemove.append(listeFichier[i])
        for i in range (0,len(toRemove)):
            listeFichier.remove(toRemove[i])
    
        print("Voici la liste des images manipulables:\n")
        for i in range (0,len(listeFichier)):
            print(i+1,":",listeFichier[i],"\n")
        while True:
            try:
                sel = int(input("Quelle image voulez vous traiter (Entrez l'indice)? "))
                im = Image.open(listeFichier[sel-1])
                break
            except:
                print("Erreur, réessayer")
                continue
        break
    while True:
        seuil = int(input("Avec quel seuil de noir et blanc souhaitez vous convertir l'image ?"))
        if seuil < 0 or seuil > 255:
            print("Ce seuil n'est pas accepté !")
            continue
        else:
            break
            
    im = passage_noir_blanc(im,seuil)
    
    im.show()
    
    sauver = input("Tapez 'oui' si vous voulez sauvegarder l'image: ")
    if sauver == 'oui':
        im.save("{} Noir & Blanc.png".format(listeFichier[sel-1][:-4]),"png")
    
    continuer = input("Tapez 'oui' si vous voulez traiter une autre image: ")
    if continuer != 'oui':
        break
```

## 3.3. Ecrire une fonction qui permet la pixellisation d’une image

```python
import os

while True:
    
    if os.getcwd()[-6:] != 'Images':
        try:
            os.chdir(os.getcwd()+'\\Images')
            break
        except FileNotFoundError:
            os.mkdir('Images')
    else:
        break

from PIL import Image
import math

def pixelliser_image(im,taille):

    larg, haut = im.size

    imPix = Image.new("RGB",(larg,haut),"black")
    borne = math.ceil((taille-1)/2)
    grosPixel={}
    for y in range (borne,haut,taille-1):
        for x in range (borne,larg,taille-1):
                r,g,b = 0,0,0
                grosPixel[x,y]={}
                for y2 in range (y-borne,y+borne):
                    for x2 in range (x-borne,x+borne):
                        try:
                            grosPixel[x,y][x2,y2] = im.getpixel((x2,y2))
                        except IndexError:
                            pass
                for item in grosPixel[x,y]:
                    r += grosPixel[x,y][item][0]
                    g += grosPixel[x,y][item][1]
                    b += grosPixel[x,y][item][2]
                r = round(r/(taille**2))
                g = round(g/(taille**2))
                b = round(b/(taille**2))
                for y2 in range (y-borne,y+borne):
                    for x2 in range (x-borne,x+borne):
                        try:
                            imPix.putpixel((x2,y2),(r,g,b))
                        except IndexError:
                            pass
                    
    return(imPix)

while True:
    while True:
        
        listeFichier = os.listdir()
        toRemove=[]
        for i in range (0,len(listeFichier)):
            if listeFichier[i][-4:] != '.jpg':
                toRemove.append(listeFichier[i])
        for i in range (0,len(toRemove)):
            listeFichier.remove(toRemove[i])
    
        print("Voici la liste des images manipulables:\n")
        for i in range (0,len(listeFichier)):
            print(i+1,":",listeFichier[i],"\n")
        while True:
            try:
                sel = int(input("Quelle image voulez vous traiter (Entrez l'indice)? "))
                im = Image.open(listeFichier[sel-1])
                break
            except:
                print("Erreur, réessayer")
                continue
        break
    larg,haut = im.size
    while True:
        try:
            taille = int(input("Avec quelle taille de pixel souhaitez vous convertir l'image ?"))
            if taille < 0 or taille > larg:
                print("Cette taille de pixel n'est pas accepté !")
                continue
        except:
            print("Erreur")
            continue
        break
            
    im = pixelliser_image(im,taille)
    
    im.show()
    
    sauver = input("Tapez 'oui' si vous voulez sauvegarder l'image: ")
    if sauver == 'oui':
        im.save("{} Pixellisé.png".format(listeFichier[sel-1][:-4]),"png")
        
    continuer = input("Tapez 'oui' si vous voulez traiter une autre image: ")
    if continuer != 'oui':
        break
```

## 3.4 Ecrire une fonction qui permet le floutage d’une image

```python
import os

while True:
    
    if os.getcwd()[-6:] != 'Images':
        try:
            os.chdir(os.getcwd()+'\\Images')
            break
        except FileNotFoundError:
            os.mkdir('Images')
    else:
        break

from PIL import Image

def flouter_image (im,taille):
    larg, haut = im.size
    borne = round((taille-1)/2)
    imFlou = Image.new("RGB",(larg,haut),"black")
    
    for y in range (0,haut):
        for x in range (0,larg):
            autourPixel={}
            r,g,b = 0,0,0
            for y2 in range (y-borne,y+borne):
                for x2 in range (x-borne,x+borne):
                    try :
                        autourPixel[x2,y2] = im.getpixel ((x2,y2))
                    except IndexError:
                            pass
            for item in autourPixel:
                r += autourPixel[item][0]
                g += autourPixel[item][1]
                b += autourPixel[item][2]
            r = round(r/(taille**2))
            g = round(g/(taille**2))
            b = round(b/(taille**2))
            try:
                imFlou.putpixel((x,y),(r,g,b)) 
            except IndexError:
                pass
    return(imFlou)

while True:
    while True:
        
        listeFichier = os.listdir()
        toRemove=[]
        for i in range (0,len(listeFichier)):
            if listeFichier[i][-4:] != '.jpg':
                toRemove.append(listeFichier[i])
        for i in range (0,len(toRemove)):
            listeFichier.remove(toRemove[i])
    
        print("Voici la liste des images manipulables:\n")
        for i in range (0,len(listeFichier)):
            print(i+1,":",listeFichier[i],"\n")
        while True:
            try:
                sel = int(input("Quelle image voulez vous traiter (Entrez l'indice)? "))
                im = Image.open(listeFichier[sel-1])
                break
            except:
                print("Erreur, réessayer")
                continue
        break
    larg,haut = im.size
    while True:
        try:
            taille = int(input("Avec quelle amplitude de flou souhaitez vous convertir l'image ? (l'execution peut prendre plusieurs minutes) "))
            if taille < 0 or taille > larg:
                print("Cette amplitude de flou n'est pas accepté !")
                continue
        except:
            print("Erreur")
            continue
        break
            
    im = flouter_image(im,taille)
    
    im.show()
    
    sauver = input("Tapez 'oui' si vous voulez sauvegarder l'image: ")
    if sauver == 'oui':
        im.save("{} Floutté.png".format(listeFichier[sel-1][:-4]),"png")
    
    continuer = input("Tapez 'oui' si vous voulez traiter une autre image: ")
    if continuer != 'oui':
        break
```
