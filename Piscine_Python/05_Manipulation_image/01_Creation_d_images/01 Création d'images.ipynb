## 2.1

```python
import os

while True:
    
    if os.getcwd()[-6:] != 'Images':
        try:
            os.chdir(os.getcwd()+'\\Images')
            break
        except FileNotFoundError:
            os.mkdir('Images')
    else:
        break

from PIL import Image

def placer_pixel (position):
    for i in range (0,len(position),2):
        im.putpixel((position[i],position[i+1]),(0,0,0))
    
im = Image.new("RGB",(800,600),"blue")

repetition = int(input("Combien de pixels voulez vous ajouter ?"))
repetition *= 2
pixel = 1
position = []
for i in range (0,repetition,2):
    x = int(input("Entrez la position x de votre pixel n° {}:".format(pixel)))
    position.append(x)
    y = int(input("Entrez la position y de votre pixel n° {}:".format(pixel)))
    position.append(y)
    pixel += 1
    
placer_pixel(position)    
im.show()
im.save("essai.png","PNG")
```

## 2.8

Explications car j'ai vraiment bidouillé (et mon cerveau à maintenant très mal):

f(x) = sqrt(rayon² - x²) fait un beau demi cercle entre x = -rayon et x = rayon.
f(x) = -sqrt(rayon² - x²) fait l'autre parti du cercle.

Problème: la conversion en entier fait perdre de l'information et sacade le cercle quand x approche de sqrt(rayon) & -x approche de -(sqrt(rayon)) .

Solution: on multiplie le rayon et x par 1000 pour le calcul, puis on divise f(x) par 1000 pour le passage à l'addition de pixel.

Inconvénient: le code met 3 bonnes secondes à s'executer. On peut multiplier x et rayon par 100 au lieu de 1000 pour gagner en rapidité, puis combler manuellement les quelques pixels manquant. 

J'ai le droit à un cadeau pour avoir réussit le défi ? :)

```python
import os

while True:
    
    if os.getcwd()[-6:] != 'Images':
        try:
            os.chdir(os.getcwd()+'\\Images')
            break
        except FileNotFoundError:
            os.mkdir('Images')
    else:
        break

from PIL import Image
from math import *

im = Image.new("RGB",(800,600),"lightgrey")

for x in range (-200000,200001): # Demi cercle inférieur
    y = (sqrt(40000000000-(x)**2))
    y = round(y/1000)
    x = round(x/1000)
    im.putpixel((x+400,y+300),(255,0,0)) #On incrémente x de 400 et y de 300 pour placer au centre de l'image
    
for x in range (-200000,200001): # Demi cercle supérieur
    y = -(sqrt(40000000000-(x)**2))
    y = round(y/1000)
    x = round(x/1000)
    im.putpixel((x+400,y+300),(255,0,0))    

im.show()
im.save("essai.png","PNG")   
```

```python
import os

while True:
    
    if os.getcwd()[-6:] != 'Images':
        try:
            os.chdir(os.getcwd()+'\\Images')
            break
        except FileNotFoundError:
            os.mkdir('Images')
    else:
        break

from PIL import Image

im = Image.new("RGB",(600,400),"white")

b = 1
for y in range (0,200):
    for x in range (0,b):
        im.putpixel((x,y),(0,0,255))
    b += 1
    if y%3 == 0:
        b += 1
        
for y in range (200,400):
    for x in range (0,b):
        im.putpixel((x,y),(0,0,255))
    for x in range (b,600):
        im.putpixel((x,y),(255,0,0))
    b -= 1
    if y%3 == 0:
        b -= 1
    
    
    
im.show()
im.save("République_Tchèque.png","PNG")
```
