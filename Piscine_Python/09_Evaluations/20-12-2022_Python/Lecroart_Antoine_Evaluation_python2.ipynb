# Exercice 1 : Température

```python
def indicateur_temperature(temperature):
    
    if temperature <= -40:
        return (-1)

    elif temperature > -40 and temperature <= -20:
        return (0)

    elif temperature > -20 and temperature <= -5:
        return (1)

    elif temperature > -5 and temperature <= 5:
        return (2)

    elif temperature > 5 and temperature <= 12:
        return (3)

    elif temperature > 12 and temperature <= 20:
        return (4)

    elif temperature > 20 and temperature <= 50:
        return (5)

    elif temperature > 50:
        return (-1)

def menu():
    while True:
        try:
            entree = float(input("Entrez une température:\n"))

        except ValueError:
            print("Entrez une température valide !\n")
            continue


        print(f"L'indicateur de température est : {indicateur_temperature(entree)}\n")


        entree = input("Voulez-vous continuer ? (yes)\n")
        if entree.find('yes') == -1:
            break
            
menu()
```

# Exercice 2 : compte_lettres

```python
def compte_lettres(mot, lettre):
    
    return (mot.count(lettre))



def menu():
    while True:
        
        while True:
            mot = input("Entrez un mot:\n")

            if mot.find(' ') != -1:
                print("Entrez un seul mot !\n")
                continue
                
            else: 
                break
           
        while True:
            lettre = input("Entrez la lettre à trouver:\n")

            if len(lettre) != 1:
                print("Saisie invalide, recommencez !\n")
                continue
                
            else:
                break

        
        print(f"'{mot}' contient {compte_lettres(mot, lettre)} fois la lettre '{lettre}'\n")


        entree = input("Voulez-vous continuer ? (yes)\n")
        if entree.find('yes') == -1:
            break
        
menu()
```

# Exercice 3 : somme paire

```python
def somme_paire(liste):
    
    if sum(liste) % 2 == 0:
        return (True)
    
    else:
        return (False)
    
    
def menu():
    while True:
        
        liste = []
        
        while True:
            try:
                entree = int(input("Entrez un entier à ajouter à la liste (autre chose pour stopper):\n"))

            except ValueError:
                break
            
            liste.append(entree)

        paire_print = ["paire", "impaire"]
        
        print(f"La somme des nombres {liste} est {paire_print[0] if somme_paire(liste) else paire_print[1]}.\n")


        entree = input("Voulez-vous continuer ? (yes)\n")
        if entree.find('yes') == -1:
            break
            
menu()
```

# Exercice 4 : gestion de dictionnaire

```python
personnes = {
    "personne1":{
        "age":18,
         "taille":180,
         "poids":80,
         "couleur_yeux":"bleus",
         "metier":"enseignant"
    },
    
    "personne2":{
         "age":87,
         "taille":190,
         "poids":59,
         "couleur_yeux":"rouges",
         "metier":"vampire"
    }
    
}

def analyse_personne(dico):

    statistiques = {}
    
    somme_age = 0
    somme_taille = 0
    somme_poids = 0
    statistiques['liste_couleurs'] = []
    statistiques['liste_metiers'] = []
    statistiques['liste_IMC'] = []
    
    
    for cle in dico:
        
        dico[cle]["IMC"] = round(dico[cle]["poids"] / ((dico[cle]["taille"] / 100)**2), 1)
        
        dico[cle]["couleur_yeux"] = dico[cle]["couleur_yeux"][0].upper()
        
        
        
        somme_age += dico[cle]['age']
        somme_taille += dico[cle]['taille']
        somme_poids += dico[cle]['poids']
        
        statistiques['liste_couleurs'].append(dico[cle]["couleur_yeux"])
        statistiques['liste_metiers'].append(dico[cle]["metier"])
        statistiques['liste_IMC'].append(dico[cle]["IMC"])
        
    statistiques['moyenne_ages'] = round(somme_age / len(dico), 1)
    statistiques['moyenne_tailles'] = round(somme_taille / len(dico))
    statistiques['moyenne_poids'] = round(somme_poids / len(dico), 1)
    
    dico['statistiques'] = statistiques
    
    return dico
        
        
analyse_personne(personnes)
```
