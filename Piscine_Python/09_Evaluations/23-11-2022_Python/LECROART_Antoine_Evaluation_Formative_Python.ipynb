# Exercice 1 : min max</br>
Le but de cet exercice est de créer une fonction min_max qui pour deux entiers a et b renvoie le couple (min,max) où min est le minimum de a et b et max le maximum.</br>
Attention, si l’un des nombres ou les deux nombres entrés sont négatifs ou nuls, la fonction renvoie l’un des message d’erreur suivant:</br>
    -‘Attention: l’un des nombres entré est négatif ou nul’ </br>
    -‘Attention:les deux nombres entrés sont négatifs ou nuls’</br>
Entrée : Deux nombresentiers a et b.</br>
Sortie : La fonction min_max doit renvoyer en résultat le couple(min,max) où min est le minimum de a et b et max le maximum.

```python
def min_max(a, b):
    
    if a <= 0 and b > 0:
        result = "Attention: l'un des nombres entré est négatif ou nuls\n"
        return(result)
        
    elif b <= 0 and a > 0:
        result = "Attention: l'un des nombres entré est négatif ou nuls\n"
        return(result)
        
    elif a <= 0 and b <= 0:
        result = "Attention: les deux nombres entré sont négatif ou nuls\n"
        return(result)
    
    else:
        if a > b:
             return((b, a))
        else:
             return((a, b))
            
while True:
    
    try:
        a = int(input("Entrez un nombre: \n"))
        b = int(input("Entrez un deuxième nombre: \n"))
        
        result = min_max(a, b)
        
        if type(result) is tuple:
            print(f"Le minimum est: {result[0]}\nLe maximum est : {result[1]}\n")
            break
            
        else:
            print(result)  
            
    except:
        print('Erreur de saisie, réessayez\n')
```

# Exercice 2: pluriel 
Le but de cet exercice est de créer une fonction pluriel qui rajoute un s au mot donné en entrée s'il ne se termine pas déjà par s.</br> 
On ne s'intéresse pas aux exceptions comme chou, oeil, cheval...</br>
Entrée : Un mot normalement mettable au pluriel.</br> 
Sortie : La fonction pluriel doit renvoyer le mot rentré avec un ‘s’ en plus à la fin s'il ne se termine pas déjà par s.

```python
def pluriel(mot):
    
    if mot[-1] != 's':
        mot += 's'
    
    return(mot)


mot = input("Entrez un mot à mettre au pluriel:\n")

print(pluriel(mot))
```

# Exercice 3: multiple de 7. 
Le but de cet exercice est de créer une fonction contient_x7 qui renvoie True si la liste donnée en entrée contient un multiple de 7 et False sinon.</br>
Entrée : Une liste de nombres.</br>
Sortie : True si la liste contient un multiple de 7. False sinon.

```python
def contient_x7(liste):
    
    result = False

    for nombre in liste:
        
        if nombre % 7 == 0 and nombre != 0:
            result = True
            
    return(result)


liste = []

while True:
    
    try:
        liste.append(int(input("Entrée un nombre à ajouter à la liste \n")))
        
    except:
        break
        
print(f"Votre liste contient un multiple de 7: {contient_x7(liste)}")
```

# Exercice 4: gestion de dictionnaire.
Le but de cet exercice est de créer une fonction gest_dico qui prend en paramètre une liste de dictionnaires présentant les caractéristiques d’une personne[{nom:ARNADI, prenom:alexandre, age:24, classe:IA1}]et un seuil d’âge.

## 4.1 : 
La fonction analyse la liste de dictionnaires et supprime tous les
dictionnaires donc la valeur ‘age’ est supérieure au seuil passé en
paramètre.</br> 
La fonction renvoie la liste de dictionnaires mise à jour.</br> 

Entrée :</br> 
liste = [{nom : ARNADI, prenom : alexandre, age : 24, classe : IA1},
{nom : PIEDO, prenom : Arthur, age : 26, classe : IA2}, {nom : MAGNO, prenom :
julien, age : 21, classe : IA1}, {nom : BERNARD, prenom : amandine, age : 31,
classe : IA1}, {nom : ARMANDI, prenom : jerome, age : 19, classe : IA2}, {nom :
BRETON, prenom : théo, age : 20, classe : IA2}]

Sortie :</br> 
liste_dico_updated = gest_dico(liste, 25)</br> 
print(liste_dico_updated)</br> 
[{nom : ARNADI, prenom : alexandre, age : 24, classe : IA1}, {nom : MAGNO,
prenom : julien, age : 21, classe : IA1}, {nom : ARMANDI, prenom : jerome, age : 19,
classe : IA2}, {nom : BRETON, prenom : théo, age : 20, classe : IA2}]

```python
def gest_dico(liste, seuil):
    
    result = []
    
    for dico in liste:
        
        if dico["age"] <= seuil:
            result.append(dico)
        
    return(result)
        
liste = [
    {"nom" : "ARNADI", "prenom" : "alexandre", "age" : 24, "classe" : "IA1"}, 
    {"nom" : "PIEDO", "prenom" : "Arthur", "age" : 26, "classe" : "IA2"}, 
    {"nom" : "MAGNO", "prenom" : "julien", "age" : 21, "classe" : "IA1"},
    {"nom" : "BERNARD", "prenom" : "amandine", "age" : 31, "classe" : "IA1"}, 
    {"nom" : "ARMANDI", "prenom" : "jerome", "age" : 19, "classe" : "IA2"}, 
    {"nom" : "BRETON", "prenom" : "théo", "age" : 20, "classe" : "IA2"}
]

liste_dico_updated = gest_dico(liste, 25)

print(liste_dico_updated)
```

## 4-2 :

La fonction sépare les dictionnaires selon la classe (IA1 ou IA2).</br> 
Elle renvoie donc deux listes de dictionnaires, l’une regroupant les dictionnaires des personnes étant en IA1 et l’autre en IA2.</br> 
Le dernier élément de chaque liste doit être un dictionnaire contenant la moyenne de tous les âges, l’âge min et l’âge max de chaque liste de dictionnaires.</br>

Entrée:</br>
liste = [{nom : ARNADI, prenom : alexandre, age : 24, classe : IA1},
{nom : PIEDO, prenom : Arthur, age : 26, classe : IA2}, {nom : MAGNO, prenom :
julien, age : 21, classe : IA1}, {nom : BERNARD, prenom : amandine, age : 31,
classe : IA1}, {nom : ARMANDI, prenom : jerome, age : 19, classe : IA2}, {nom :
BRETON, prenom : théo, age : 20, classe : IA2}]

Sortie:</br>
IA1, IA2 = gest_dico(liste)</br>
print(IA1)</br>
[{nom : ARNADI, prenom : alexandre, age : 24, classe : IA1}, {nom : MAGNO,
prenom : julien, age : 21, classe : IA1}, {nom : BERNARD, prenom : amandine, age :
31, classe : IA1}, {moyenne :25.3, min :21, max :31}]</br>
print(IA2) </br>
[{nom : PIEDO, prenom : Arthur, age : 26, classe : IA2}, {nom : ARMANDI, prenom :
jerome, age : 19, classe : IA2}, {nom : BRETON, prenom : théo, age : 20, classe :
IA2}, {moyenne :21.6, min :19, max :26}]

```python
def gest_dico_bis(liste):
    
    IA1, IA2 = [], []
    
    for dico in liste:
        
        if dico["classe"] == 'IA1':
            IA1.append(dico)
            
        elif dico["classe"] == 'IA2':
            IA2.append(dico)
    
    liste_temp = []
    moyenne = 0
    
    for dico in IA1:
        
        moyenne += dico["age"]
        liste_temp.append(dico["age"])
        
    liste_temp.sort()
    moyenne /= len(IA1)
    
    IA1.append({"moyenne" : round(moyenne, 1), "min" : liste_temp[0], "max" : liste_temp[-1]})
    
    
    
    liste_temp = []
    moyenne = 0
    
    for dico in IA2:
        
        moyenne += dico["age"]
        liste_temp.append(dico["age"])
        
    liste_temp.sort()
    moyenne /= len(IA2)
    
    IA2.append({"moyenne" : round(moyenne, 1), "min" : liste_temp[0], "max" : liste_temp[-1]})
    
    return(IA1, IA2)
    
    
liste = [
    {"nom" : "ARNADI", "prenom" : "alexandre", "age" : 24, "classe" : "IA1"}, 
    {"nom" : "PIEDO", "prenom" : "Arthur", "age" : 26, "classe" : "IA2"}, 
    {"nom" : "MAGNO", "prenom" : "julien", "age" : 21, "classe" : "IA1"},
    {"nom" : "BERNARD", "prenom" : "amandine", "age" : 31, "classe" : "IA1"}, 
    {"nom" : "ARMANDI", "prenom" : "jerome", "age" : 19, "classe" : "IA2"}, 
    {"nom" : "BRETON", "prenom" : "théo", "age" : 20, "classe" : "IA2"}
]

IA1, IA2 = gest_dico_bis(liste)
print(f"Promo IA1:\n{IA1}\n\nPromo IA2:\n{IA2}")
```
